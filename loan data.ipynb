```python
import pandas as pd
Covid_data=pd.read_csv("loan_data.csv")
Covid_data.isnull().sum()

Covid_data.fillna(method="ffill",inplace=True)
Covid_data.fillna(method="bfill",inplace=True)

Covid_data1 = pd.get_dummies (Covid_data, columns=["Gender","Married","Dependents","Education","Self_Employed","Property_Area","Loan_Status"])

Covid_data1.dtypes
Covid_data1.drop("Gender_Female",axis=1,inplace=True)
Covid_data1.drop("Married_No",axis=1,inplace=True)
Covid_data1.drop("Education_Graduate",axis=1,inplace=True)
Covid_data1.drop("Self_Employed_No",axis=1,inplace=True)
Covid_data1.drop("Property_Area_Rural",axis=1,inplace=True)
Covid_data1.drop("Loan_Status_N",axis=1,inplace=True)
Covid_data1.drop("Dependents_3+",axis=1,inplace=True)


x=Covid_data1.iloc[:,2:15]
y=Covid_data1.iloc[:,-1]

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=(0))

from sklearn.linear_model import LogisticRegression
m1=LogisticRegression()
m1.fit(x_train, y_train)
y_pred=m1.predict(x_test)

from sklearn.metrics import accuracy_score,confusion_matrix
acc=accuracy_score(y_test, y_pred)
c=confusion_matrix(y_test,y_pred)
print(acc)
print(c)
```
